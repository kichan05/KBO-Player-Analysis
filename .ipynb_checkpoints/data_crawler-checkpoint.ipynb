# 데이터베이스 관리 함수

```python
conn = sqlite3.connect('baseball.db')
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS hitter (
        player_id INTEGER PRIMARY KEY,
        "타율" REAL,
        "경기" INTEGER,
        "타석" INTEGER,
        "타수" INTEGER,
        "득점" INTEGER,
        "안타" INTEGER,
        "2루타" INTEGER,
        "3루타" INTEGER,
        "홈런" INTEGER,
        "루타" INTEGER,
        "타점" INTEGER,
        "도루" INTEGER,
        "도루실패" INTEGER,
        "희생번트" INTEGER,
        "희생플라이" INTEGER,
        "볼넷" INTEGER,
        "고의4구" INTEGER,
        "사구" INTEGER,
        "삼진" INTEGER,
        "병살타" INTEGER,
        "장타율" REAL,
        "출루율" REAL,
        "실책" INTEGER,
        "도루성공률" REAL,
        "멀티히트" INTEGER,
        "출루율+장타율" REAL,
        "득점권타율" REAL,
        "대타타율" REAL
    )
""")

cur.execute("""
    CREATE TABLE IF NOT EXISTS pitcher (
        player_id INTEGER PRIMARY KEY,
        "평균자책점" REAL,
        "경기" INTEGER,
        "완투" INTEGER,
        "완봉" INTEGER,
        "승리" INTEGER,
        "패배" INTEGER,
        "세이브" INTEGER,
        "홀드" INTEGER,
        "승률" REAL,
        "타자수" INTEGER,
        "투구수" INTEGER,
        "이닝" REAL,
        "피안타" INTEGER,
        "2루타" INTEGER,
        "3루타" INTEGER,
        "홈런" INTEGER,
        "희생번트" INTEGER,
        "희생플라이" INTEGER,
        "볼넷" INTEGER,
        "고의4구" INTEGER,
        "삼진" INTEGER,
        "폭투" INTEGER,
        "보크" INTEGER,
        "실점" INTEGER,
        "자책점" INTEGER,
        "블론세이브" INTEGER,
        "이닝당 출루허용률" REAL,
        "피안타율" REAL,
        "퀄리티스타트" INTEGER
    )
""")

cur.execute("""
    CREATE TABLE IF NOT EXISTS player (
        player_id INTEGER PRIMARY KEY,
        uniform_number INTEGER NULL,
        name TEXT,
        team TEXT,
        position TEXT,
        birth_day TEXT,
        height INTEGER,
        weight INTEGER,
        school TEXT
    )
""")
```

```python
def insert_hitter(data):
    global cur, conn
    
    columns = ', '.join(f'"{k}"' for k in data.keys())  # "타율", "경기", ...
    placeholders = ', '.join('?' for _ in data)        # ?, ?, ...
    sql = f"INSERT INTO hitter ({columns}) VALUES ({placeholders})"

    cur.execute(sql, tuple(data.values()))
    conn.commit()

def insert_pitcher(data):
    global cur, conn
    
    columns = ', '.join(f'"{k}"' for k in data.keys()) # "평균자책점", "경기"...
    placeholders = ', '.join('?' for _ in data)       # ?, ?, ? ...
    
    sql = f"INSERT INTO pitcher ({columns}) VALUES ({placeholders})"
    
    # 5. 실행 및 저장
    cur.execute(sql, tuple(data.values()))
    conn.commit()

def insert_player(data):
    global conn, cur

    try:
        columns = ', '.join(data.keys())
        placeholders = ', '.join('?' for _ in data)
        sql = f"INSERT OR REPLACE INTO player ({columns}) VALUES ({placeholders})"
        
        cur.execute(sql, tuple(data.values()))
        conn.commit()
        print(f"✅ 저장 완료: {data.get('name')} (ID: {data.get('player_id')})")
    except Exception as e:
        print(f"❌ 에러 발생: {e}")
```

```python
def close_database():
    global cur, conn
    conn.close()
```

```python
close_database()
```

```python
def get_pagenation_size():
    pagenation_element = driver.find_element(By.CSS_SELECTOR, '.paging')
    page_count = len(pagenation_element.find_elements(By.CSS_SELECTOR, "& > *")) - 2
    return page_count
```

```python
driver = webdriver.Chrome()
driver.get("https://www.koreabaseball.com/Player/Search.aspx")
```

```python
def foramt_default_player_info(row):
    column_list = row.select("td")
    
    player_id = row.select_one("a")["href"].split("playerId=")[1]
    uniform_number = column_list[0].text
    name = column_list[1].text
    team = column_list[2].text
    position = column_list[3].text
    birth_day = column_list[4].text
    school = column_list[6].text
    
    phisical = column_list[5].text
    pattern = r"(\d+)cm,\s*(\d+)kg"
    match = re.search(pattern, phisical)

    if match:
        height = match.group(1)
        weight = match.group(2)
    else:
        height = None
        weight = None


    return {
        "player_id" : player_id,
        "uniform_number" : uniform_number,
        "name" : name,
        "team" : team,
        "position" : position,
        "birth_day" : birth_day,
        "height" : height,
        "weight" : weight,
        "school" : school,
    }
```

```python
from fractions import Fraction
```

```python
def to_baseball_notation(text):
    """
    '160 1/3' -> 160.1 (실수형)
    '160 2/3' -> 160.2 (실수형)
    """
    # 1. 텍스트가 비었는지 방어
    if not text: return 0.0
    
    # 2. 문자열 자체를 치환해버리는 게 가장 Sexy합니다.
    # (공백 주의: ' 1/3'을 '.1'로 바꿉니다)
    text = text.replace(' 1/3', '.1').replace(' 2/3', '.2')
    
    return float(text)
```

```python
def get_player_detail_info():
    global driver
    html = driver.page_source
    soup = BeautifulSoup(html)
    result = {}


    table_head = []
    table_body = []
    table = soup.select(".tbl.tt")
    

    for n, t in enumerate(table):
        if(n == 2):
            break

        if(len(t.select("td")) == 1):
            return {}
        
        table_head.extend(t.select("th"))
        table_body.extend(t.select("td"))
    
    for th, td in zip(table_head, table_body):
        if th.text == "팀명":
            continue
    
        key = th.select_one("a")["title"]

        if(key == "이닝"):
            raw_ip = td.text
            real_ip = float(sum(map(Fraction, raw_ip.split())))
            result[key] = real_ip
        else:
            result[key] = literal_eval("0" if td.text == "-" else td.text)
    
    return result
```

## 선수 세부 정보 크롤링

```python
df_player_defualt_info = pd.read_csv("plyer_default_info.csv")
df_player_defualt_info.head()
```

```python
for row in tqdm(df_player_defualt_info.itertuples(), total = df_player_defualt_info.shape[0], desc="데이터 크롤링중"):
    player_id = row.player_id
    position = row.position
    if(position == "투수"):
        url = f"https://www.koreabaseball.com/Record/Player/PitcherDetail/Basic.aspx?playerId={player_id}"
    else:
        url = f"https://www.koreabaseball.com/Record/Player/HitterDetail/Basic.aspx?playerId={player_id}"
    driver.get(url)

    sleep()

    res = get_player_detail_info()
    if res:
        res["player_id"] = player_id
        if(position == "투수"):
            insert_pitcher(res)
        else:
            insert_hitter(res)
    if((row.Index + 1) % 10 == 0):
        sleep(5)
    elif((row.Index + 1) % 5 == 0):
        sleep(1)
    else:
        sleep()
```
