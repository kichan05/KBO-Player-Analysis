# ðŸ”– PCA

```python
import pandas as pd
import numpy as np
import seaborn as sns; import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df.columns
```

## 1. PCA fitting

ðŸ‘‰ ìˆ«ìžë¡œ ëœ ë³€ìˆ˜ë“¤ ì„ íƒ

```python
df_X_cts = df[['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
              'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
              'pH', 'sulphates', 'alcohol']]
```

ðŸ‘‰ í‘œì¤€í™”(standardization)

```python
scaler = StandardScaler()
df_X_cts_standardized = scaler.fit_transform(df_X_cts)
```

ðŸ‘‰ ë³€ìˆ˜ ê°œìˆ˜ë§Œí¼ ì£¼ì„±ë¶„ ê°œìˆ˜ ì§€ì •í•´ PCA ìˆ˜í–‰

```python
num_components = 11
pca = PCA(num_components)
PCA_result = pca.fit_transform(df_X_cts_standardized)
```

## 2. PCA results

ðŸ‘‰ ê° ì£¼ì„±ë¶„ì´ ì„¤ëª…í•˜ëŠ” variance (loadings)

```python
print(pca.explained_variance_ratio_)
```

```python
sum(pca.explained_variance_ratio_)
```

ðŸ‘‰ scree plot

```python
plt.plot(np.arange(1, num_components + 1),
         pca.explained_variance_,
         marker='o', linestyle='-', color='blue')
plt.xlabel("the number of components")
plt.ylabel("eigenvalue")
```

```python
pca.explained_variance_ratio_[:3]
```

```python
sum(pca.explained_variance_ratio_[:3])
```

ðŸ‘‰ ìƒˆë¡œ êµ¬ì„±ëœ ì£¼ì„±ë¶„ì˜ ê°’ë“¤

```python
pd.DataFrame(PCA_result)
```

ðŸ‘‰ ì£¼ì„±ë¶„-ë³€ìˆ˜ ìƒê´€ê³„ìˆ˜ (Loadings)

```python
correlation_XZ1 = [float(np.corrcoef(PCA_result[:,0], df_X_cts_standardized[:,i])[0,1]) for i in range(df_X_cts_standardized.shape[1])]
correlation_XZ2 = [float(np.corrcoef(PCA_result[:,1], df_X_cts_standardized[:,i])[0,1]) for i in range(df_X_cts_standardized.shape[1])]
correlation_XZ1Z2 = pd.DataFrame({"Z1_correlation":correlation_XZ1,
                                 "Z2_correlation":correlation_XZ2},
                                index = df_X_cts.columns)
correlation_XZ1Z2
```

## 3. ì‹œê°í™”

ðŸ‘‰ íŠ¹ì„±ë³„ ê±°ë¦¬

```python
x = correlation_XZ1Z2['Z1_correlation']
y = correlation_XZ1Z2['Z2_correlation']

plt.scatter(x, y, color="black")

for i, txt in enumerate(correlation_XZ1Z2.index):
    plt.annotate(txt, (x[i]+0.02, y[i]+ 0.05), fontsize=8)

plt.axvline(0); plt.axhline(0)

plt.show()
```

ðŸ‘‰ ì œí’ˆì˜ ìœ„ì¹˜

```python
plt.scatter(PCA_result[:,0], PCA_result[:,1], marker="+")

for i in range(PCA_result.shape[0]):
    plt.annotate(i, (PCA_result[i,0]+0.1, PCA_result[i,1]+ 0.1), fontsize=5)
```

ðŸ‘‰ ì ˆëŒ€ì ìœ¼ë¡œ í‰ê· ì¹˜ì— í•´ë‹¹í•˜ëŠ” ì œí’ˆì€ ì–´ëŠ ìœ„ì¹˜ì— ìžˆì„ê¹Œ?

```python
# If all values are mean,
z_mean = scaler.transform(np.zeros((1, 11)))

plt.scatter(PCA_result[:,0], PCA_result[:,1])
plt.scatter(z_mean[0,0], z_mean[0,1], color="red")
```

ðŸ‘‰ ë ˆë“œì™€ì¸ vs í™”ì´íŠ¸ì™€ì¸

```python
# notice that we set up lists to track each group's plotting color and label
colors = ['r','c','b']
label_text = ["Reds", "Whites"]

# and we loop over the different groups
df['color_group'] = np.where(df['color']=='red', 0, 1)

plt.scatter(PCA_result[df['color_group']==0, 0],
            PCA_result[df['color_group']==0,1],
            color="Red", label="Red Wine")
plt.scatter(PCA_result[df['color_group']==1, 0],
            PCA_result[df['color_group']==1,1],
            color="Blue", label="White Wine")
plt.legend()
plt.show()
```

ðŸ‘‰ ì£¼ì„±ë¶„ê³¼ ë‹¤ì¤‘ê³µì„ ì„±

```python
import statsmodels.api as sm
```

```python
y = df['quality']
X = sm.add_constant(df_X_cts)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_df = pd.DataFrame()
vif_df['IVs'] = df_X_cts.columns
vif_df["VIF"] = [variance_inflation_factor(df_X_cts.values, k) for k in range(df_X_cts.shape[1])]

vif_df
```

```python
y = df['quality']
X = sm.add_constant(pd.DataFrame(PCA_result))
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_df = pd.DataFrame()
vif_df['IVs'] = pd.DataFrame(PCA_result).columns
vif_df["VIF"] = [variance_inflation_factor(pd.DataFrame(PCA_result).values, k) for k in range(pd.DataFrame(PCA_result).shape[1])]

vif_df
```
