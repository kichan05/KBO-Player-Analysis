# üîñ Factor Analysis

```python
! pip install factor-analyzer
```

```python
import pandas as pd
import numpy as np
import seaborn as sns; import matplotlib.pyplot as plt
# from sklearn.decomposition import FactorAnalysis
import factor_analyzer # https://factor-analyzer.readthedocs.io/en/latest/factor_analyzer.html
from sklearn.preprocessing import StandardScaler
```

üëâ Ïà´ÏûêÎ°ú Îêú Î≥ÄÏàòÎì§ ÏÑ†ÌÉù

```python
df_X_cts = df[['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
              'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
              'pH', 'sulphates', 'alcohol']]
```

üëâ [Ï∞∏Í≥†Ïö©] Bartlett Sphericity Test

```python
factor_analyzer.calculate_bartlett_sphericity(df_X_cts)
```

üëâ ÌëúÏ§ÄÌôî(standardization)

```python
scaler = StandardScaler()
df_X_cts_standardized = scaler.fit_transform(df_X_cts)
```

üëâ Í∞ÄÎä•Ìïú ÏµúÎåÄ ÏöîÏù∏ ÏàòÎ°ú ÏùºÎã® fitting

```python
num_factors = min(df_X_cts_standardized.shape)-1
fa= factor_analyzer.FactorAnalyzer(num_factors, rotation='varimax')
FA_result = fa.fit(df_X_cts_standardized)
```

```python
FALoadings = pd.DataFrame(FA_result.loadings_)
FALoadings.index = df_X_cts.columns
FALoadings
```

üëâ scree plot

```python
FA_result.get_eigenvalues()[0] # 0: original eigenvalues; 1: common factor eigenvalues
```

```python
FA_result.get_eigenvalues()[1] # 0: original eigenvalues; 1: common factor eigenvalues
```

```python
pd.DataFrame({"original eigenvalue":FA_result.get_eigenvalues()[0],
              "common factor eigenvalue": FA_result.get_eigenvalues()[1]})
```

```python
plt.plot(np.arange(1, len(FA_result.get_eigenvalues()[0]) + 1),
         FA_result.get_eigenvalues()[0],
         marker='o', linestyle='-', color='blue')
plt.xlabel("the number of factor")
plt.ylabel("eigenvalue")
```

üëâ ÏöîÏù∏ Ïàò Ï°∞Ï†ï

```python
num_factors = 3
fa= factor_analyzer.FactorAnalyzer(num_factors, rotation='varimax')
FA_result = fa.fit(df_X_cts_standardized)
```

```python
Factor_scores = FA_result.fit_transform(df_X_cts_standardized)
Factor_scores = pd.DataFrame(Factor_scores)
Factor_scores
```

```python
df['Factor1'] = Factor_scores.iloc[:,0]
df['Factor2'] = Factor_scores.iloc[:,1]
x = df['Factor1']
y = df['Factor2']

plt.scatter(x, y, marker="+")

for i, txt in enumerate(df.index):
    plt.annotate(txt, (x[i]+0.1, y[i]+ 0.1), fontsize=5)

plt.axvline(0); plt.axhline(0)

plt.show()
```
